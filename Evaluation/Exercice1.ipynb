```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ebola_guinea.csv')
```

```python
# Le nombre totale de cas, de décés et de taux de mortalité par préfecture
taux_mortalite = df.groupby('Prefecture').agg({
    'Cas': 'sum',
    'Deces': 'sum'
}).assign(taux_mortalite=lambda x: (x['Deces'] / x['Cas'] * 100).round(2))  
print("\nTaux de mortalité par préfecture:")
print(taux_mortalite)
```

```python
# Graphique 1 : Nombre total de cas par préfecture
plt.figure(figsize=(10, 6))

cas_par_prefecture = df.groupby('Prefecture')['Cas'].sum()

bars = plt.bar(cas_par_prefecture.index, cas_par_prefecture.values, color='#4ECDC4')

plt.title('Nombre total de cas par préfecture', pad=15, fontsize=12, fontweight='bold')
plt.xlabel('Préfecture')
plt.ylabel('Nombre total de cas')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
